```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

tf.compat.v1.disable_eager_execution() #This line has to be there for the GPU to work on my laptop. Don't ask me why. ChatGPT came up with it when seeing my error message
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

def generate_data(n,c,mu): #Generates the data
    d=np.zeros((n,c)) #Initiallizing the matrix to 0
    a=np.random.randint(low=0,high=c,size=n) #The random index that is to be non-zero
    b=np.random.rand(n) #The random number for the non-zero index
    for i in range(n):
        d[i,a[i]]=1-mu*b[i]
    return d
def mse(data1,data2): #Evaluates the MSE between 2 datasets
    return np.average((data1-data2)**2)

samples=10000 #Set number of epochs
n=80 #Set size of dataset
c=5  #Set number of neurons in input and output layer
r=2  #Set number of neurons in intermediate layer

colors=['r','b','g','y','k'] #Colors for plotting
for mu in np.linspace(0,1,11):
    for j in range(5):
    
        data=generate_data(n,c,mu) #Generating training data
        
        
        
        
        odata=data
        tdata=generate_data(n,c,mu) #Generating test data
        otdata=tdata
        data.reshape((n,c,1))
        tdata.reshape((n,c,1))

        #Defining the model, giving it a random initiallisation
        model=tf.keras.models.Sequential([
            tf.keras.layers.Flatten(input_shape=(c,)),
            tf.keras.layers.Dense(r,activation='relu'),
            tf.keras.layers.Dense(c)])
        model.build()
        model.summary()
        
        
        model.compile(optimizer='sgd', loss='mse') #Using SGD to optimise, and using MSE as the loss
    
    
        performance=np.zeros((samples,2)) #Array to keep track of performance throughout the training
        for i in range(samples):
            model.fit(data,odata,epochs=1,verbose=0) #Training the model for 1 epoch
            pre=model.predict(data) #Prediction on the training data
            pret=model.predict(tdata) #Prediction on the test data
            performance[i,0]=mse(pre,odata) #Recordingn the MSE performance on the training set after i epochs 
            performance[i,1]=mse(pret,otdata) #Recording the MSE performance on the test set after i epochs
        print(j) #print to show how the code progresses
        plt.plot(performance[:,0],label='run '+str(j+1),color=colors[j]) #Plot the performance on the training data
        plt.plot(performance[:,1],color=colors[j],linestyle='--') #Plot the performance on the test data
    plt.legend()
    plt.yscale('log')
    plt.xlabel('Epochs')
    plt.ylabel('MSE Loss')
    plt.savefig('fTSMfig'+str(mu)+'.png')
    plt.title('$\mu \in ['+str(np.around(1-mu,decimals=1))+',1]$')
    plt.show()
    plt.clf()

```
